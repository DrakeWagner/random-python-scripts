```python
# An analysis to up my inner "90's kid" pokemon skills

import pandas as pd
import numpy as np
import scipy

poke = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv')
```

```python
poke[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].describe().round(2)
```

```python
# I only played up to gen 4, so I'm removing all the other pokemon
poke = poke[poke['Generation'].between(0,4)] \
                .rename(columns = {'Type 1':'type_1',
                'Type 2':'type_2',
                'Sp. Atk':'sp_atk',
                'Sp. Def':'sp_def'})
poke.columns = poke.columns.str.lower()
```
