```python
# minimal eda
import os
import re
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt


os.getcwd()

filelist = os.listdir()
files = []
for i in filelist:
    if i.endswith('.csv') and i not in files:
        files.append(i)
```

```python
# Assign dates
files2 = []
def AssignDates(file_list):
    for file in file_list:
        open_file = pd.read_csv(file)
        open_file['Date'] = re.split(r"_", file)[0]
        print(open_file)
        files2.append(open_file)
```
